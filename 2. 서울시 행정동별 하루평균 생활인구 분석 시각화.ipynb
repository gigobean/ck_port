### 서울 열린 데이터 광장
### 서울시 행정동별 하루평균 생활인구 분석 시각화
### KT LTE 시그널 데이터로 지역 생활인구 추정

```python
# 데이터 불러오기
import pandas as pd

df1_raw = pd.read_excel('./data/서울시_행정동별_생활인구수.xlsx')
df1 = df1_raw
df1
```

```python
# 상위 10개 행정동 확인
df1 = df1.sort_values(by = '총생활인구수', ascending = False)
df_top10 = df1.head(10)
df_top10
```

```python
# 서울시 행정동별 생활인구 분석 상위 10개 동 시각화
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 5))
plt.rc('font', size=15)
plt.title('서울시 행정동별 하루평균 생활인구 상위 10개 동')
plt.barh(df_top10['행정동명'], df_top10['총생활인구수'])
plt.show()
```

```python
# 각 구단위 생활인구수
gu_sum = df1.groupby('시군구명').sum().sort_values('총생활인구수', ascending = False)
gu_sum.reset_index(level=0, inplace=True)
gu_sum.head()
```

```python
# 서울시 각 구별 하루 평균 생활 인구수 시각화 bar
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 8))
plt.rc('font', size=15)
plt.title('서울시 각 구별 하루평균 생활인구수')
plt.barh(gu_sum['시군구명'], gu_sum['총생활인구수'])
plt.show()
```

```python
# 서울시 각 구별 하루평균 생활인구수 pie chart
plt.rc('font', size=15)
plt.figure(figsize = (10,10))
plt.title('서울시 각 구별 하루평균 생활인구수')
plt.pie(gu_sum['총생활인구수'], labels = gu_sum['시군구명'], autopct = '%.f%%')
plt.show()
```
